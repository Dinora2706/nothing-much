```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x = ["Year 1", "Year 2", "Year 3"]
y1 = np.array([200,300,500])
y2 = np.array([50, 60, 100])
y3 = np.array([30,50,100])
y4 = np.array([20, 30, 50])
y5 = np.array([100, 150, 300])
plt.bar(x, y1, color = "r")
plt.bar(x,y2, bottom = y1, color = "b")
plt.bar(x,y3, bottom = y1+y2, color = "y")
plt.bar(x, y4, bottom = y1+y2+y3, color = "g")
plt.bar(x, y5, bottom = y1+y2+y3+y4, color = "purple")
plt.xlabel("")
plt.ylabel("Number of students")
plt.legend(["English", "Korean", "Chinese", "Arabic, Spanish etc", "SAT/ACT"])
plt.title("Service usage by year")
plt.show
```
